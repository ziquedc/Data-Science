```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dados = pd.read_csv('breast-cancer.csv')
dados
```

Base de dados baixada do Keggle: https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

```python
dados.describe()
```

```python
# M => 0 
# B=> 1
dados.diagnosis.replace(['M', 'B'], [1, 0], inplace=True)
dados.drop(['id'], axis=1, inplace=True)
```

```python
# verificar balanceamento dos dados
np.unique(dados['diagnosis'], return_counts=True)
```

```python
# verificar dados faltantes
dados.isnull().sum()
```

```python
sns.countplot(x = dados['diagnosis']);
```

```python
grafico = px.scatter_matrix(dados, dimensions=['texture_mean','perimeter_mean','area_mean'], color='diagnosis');
grafico.show();
```

```python
# DIVISÃO ENTRE PREVISORES E CLASSES
X_dados    = dados.iloc[:,1:].values
Y_dados    = dados.iloc[:,0].values
```

```python
# shape das variáveis
X_dados.shape, Y_dados.shape
```

```python
# ESCALONAMNTO DE VALORES => Deixar todas as colunas na mesma escala de valores (evita que os algoritmos coloquem pesos maiores e colunas)
from sklearn. preprocessing import StandardScaler
scaler = StandardScaler()
X_dados = scaler.fit_transform(X_dados)
```

```python
from sklearn.model_selection import GridSearchCV #cross validation
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

# Tuning de parâmetros

```python
# ARVORE DE DECISÃO
parametros = { 
    'criterion' :   ['gini', 'entropy'],
    'splitter'  :   ['best','random'],
    'min_samples_split' :   [2,5,10],
    'min_samples_leaf'  :   [1,5,10],
    'max_depth'         :   [None,5,10]
}

grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=parametros)
grid_search.fit(X_dados,Y_dados)

melhores_parametros = grid_search.best_params_
melhor_resultado    = grid_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
# RANDOM FOREST
parametros = { 
    'criterion' :   ['gini', 'entropy'],
    'n_estimators' : [10,40,100,150],
    'min_samples_split' :   [2,5,10],
    'min_samples_leaf'  :   [1,5,10]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=parametros)
grid_search.fit(X_dados,Y_dados)

melhores_parametros = grid_search.best_params_
melhor_resultado    = grid_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
# KNN
parametros = { 
    'weights' :   ['uniform', 'distance'],
    'n_neighbors' : [3,5,10,20],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p':        [1,2]
}

grid_search = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parametros)
grid_search.fit(X_dados,Y_dados)

melhores_parametros = grid_search.best_params_
melhor_resultado    = grid_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
# LOGISTIC REGRESSION
parametros = { 
    'tol' :   [0.0001, 0.00001, 0.000001],
    'C' : [1.0,1.5,2.0],
    'solver' : ['lbfgs', 'sag', 'saga']
}

grid_search = GridSearchCV(estimator=LogisticRegression(), param_grid=parametros);
grid_search.fit(X_dados,Y_dados);

melhores_parametros = grid_search.best_params_;
melhor_resultado    = grid_search.best_score_;
print(melhores_parametros);
print(melhor_resultado);
```

```python
# SVM
parametros = { 
    'tol' :   [0.001, 0.0001, 0.00001],
    'C' : [1.0,1.5,2.0],
    'kernel' : ['rbf', 'linear', 'poly','sigmoid']
}

grid_search = GridSearchCV(estimator=SVC(), param_grid=parametros)
grid_search.fit(X_dados,Y_dados)

melhores_parametros = grid_search.best_params_
melhor_resultado    = grid_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
# REDE NEURAL
# Camadas rede neural => (30 + 2) / 2 = 16
# 30 = Neurônios de entrada
# 2 = Neurônios de saída
# 16 = Camadas recomendada para essa base de dados
parametros = { 
    'activation' :   ['relu', 'logistic','than'],
    'solver' : ['adam','sgd'],
    'batch_size' : [10,56],
    'hidden_layer_sizes': [(16,16)],
    'max_iter' : [200,500,1000]
}

grid_search = GridSearchCV(estimator=MLPClassifier(), param_grid=parametros)
grid_search.fit(X_dados,Y_dados)

melhores_parametros = grid_search.best_params_
melhor_resultado    = grid_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

# Validação cruzada

```python
from sklearn.model_selection import cross_val_score, KFold
resultados_arvore           = []
resultados_randomf          = []
resultados_knn              = []
resultados_logistic         = []
resultados_svm              = []
resultados_redeneural       = []

# faz 30 teste a cada K do cross validation => K=10 => 10*30 = 300 testes para cada algoritmo
# salva a média de cada teste de cada algoritmo mas respectivas variáveis

for i in range(30):
    # n_splits      => k vezes a separar os dados
    # shuffle=True  => sempre misturar os dados
    kfold = KFold(n_splits=10, shuffle=True, random_state=i) 
    
    # instaciar os algoritmos com os melhores parametros encontrados durante o Tuning
    arvore = DecisionTreeClassifier(criterion='entropy', max_depth= 10, min_samples_leaf= 1, min_samples_split=  2, splitter='random')
    scores1 = cross_val_score(arvore, X_dados, Y_dados, cv=kfold)
    resultados_arvore.append(scores1.mean())
   
    randoforest = RandomForestClassifier(criterion='entropy', min_samples_leaf= 5, min_samples_split=  5, n_estimators=150)
    scores2 = cross_val_score(randoforest, X_dados, Y_dados, cv=kfold)
    resultados_randomf.append(scores2.mean())

    knn = KNeighborsClassifier(algorithm='auto', n_neighbors=10, p=2, weights='distance')
    scores3 = cross_val_score(knn,  X_dados, Y_dados, cv=kfold)
    resultados_knn.append(scores3.mean())

    logistic = LogisticRegression(C=1.0, solver= 'lbfgs', tol=0.0001)
    scores4 = cross_val_score(logistic,  X_dados, Y_dados, cv=kfold)
    resultados_logistic.append(scores4.mean())

    svm = SVC(C=1.5, kernel= 'rbf',tol=0.0001)
    scores5 = cross_val_score(svm,  X_dados, Y_dados, cv=kfold)
    resultados_svm.append(scores5.mean())

    redeneural = MLPClassifier(activation='logistic', batch_size=10, solver='sgd', max_iter=200, hidden_layer_sizes=(16,16) ) #hidden_layer_sizes=(5,5)
    scores6 = cross_val_score(redeneural,  X_dados, Y_dados, cv=kfold)
    resultados_redeneural.append(scores6.mean())

 

    
```

```python
# ANÁLISE DOS RESULTADOS
resultados = pd.DataFrame({
    'arvore': resultados_arvore, 'random forest': resultados_randomf, 'knn': resultados_knn, 'logistic': resultados_logistic, 
    'svm': resultados_svm, 'redeneural':resultados_redeneural
})
```
